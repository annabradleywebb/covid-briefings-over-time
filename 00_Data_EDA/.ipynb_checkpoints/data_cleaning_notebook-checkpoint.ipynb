```python
from pymongo import MongoClient
from pprint import pprint
import nltk
import re
import string
import pandas as pd
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import CountVectorizer
import sklearn.feature_extraction
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from data_cleaning import cleaner, clean_content, add_word_tokenized_column, add_stem_column, stem_text, lemmatize_nouns, lemmatize_verbs, add_noun_column, add_verb_column
import pickle
from nltk.stem import WordNetLemmatizer 
from nltk.stem.lancaster import LancasterStemmer
```

```python
#Import databases from Mongo
client = MongoClient()
db = client.speeches
```

```python
collection_us = db.usa
us_df = pd.DataFrame(list(collection_us.find()))

collection_it = db.italy
it_df = pd.DataFrame(list(collection_it.find()))
```

```python
clean_content([us_df, it_df])
```

```python
#US-specific cleanup: removing documents from before March or after April
us_df['Date'] = pd.to_datetime(us_df['Date'])
mask = us_df['Date']<='2020-04-30'
us_df = us_df[mask]
mask2 = us_df['Date']>='2020-03-01'
us_df = us_df[mask2]
```

```python
pickle.dump(us_df, open("us_df.p", "wb"))
pickle.dump(it_df, open("it_df.p", "wb"))
```

```python
nltk.download('stopwords')
italian_words = set(nltk.corpus.stopwords.words('italian'))

#Taking out Mr. and President because they are very common due to reporters addressing Trump
english_words = set(nltk.corpus.stopwords.words('english'))
english_words = english_words.union(['mr', 'president'])
```

```python
#Count vectorizer with single words, English
cv_1_en = CountVectorizer(stop_words=english_words, ngram_range=(1,1))
X_en_1 = cv_1_en.fit_transform(us_df['Content'])
word_counts_en_1 = pd.DataFrame(X_en_1.toarray(),columns=cv_1_en.get_feature_names())
```

```python
#Count vectorizer with single words, Italian
cv_1_it = CountVectorizer(stop_words=italian_words, ngram_range=(1,1))
X_it_1 = cv_1_it.fit_transform(it_df['Content'])
word_counts_it_1 = pd.DataFrame(X_it_1.toarray(),columns=cv_1_it.get_feature_names())
```

```python
#Count vectorizer with two words, English
cv_2_en = CountVectorizer(stop_words=english_words, ngram_range=(2,2))
X_en_2 = cv_2_en.fit_transform(us_df['Content'])
word_counts_en_2 = pd.DataFrame(X_en_2.toarray(),columns=cv_2_en.get_feature_names())
```

```python
#Count vectorizer with two words, Italian
cv_2_it = CountVectorizer(stop_words=italian_words, ngram_range=(2,2))
X_it_2 = cv_2_it.fit_transform(it_df['Content'])
word_counts_it_2 = pd.DataFrame(X_it_2.toarray(),columns=cv_2_it.get_feature_names())
```

```python
#Saving files to use for modeling
pickle.dump(word_counts_en_1, open("en_1.p", "wb"))
pickle.dump(word_counts_it_1, open("it_1.p", "wb"))
pickle.dump(word_counts_en_2, open("en_2.p", "wb"))
pickle.dump(word_counts_it_2, open("it_2.p", "wb"))
```

```python
cv_i = CountVectorizer(stop_words=italian_words, ngram_range=(2,2))
X_i = cv_i.fit_transform(it_df['Content'])
```
